```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd
```

```python
df = pd.read_csv('./Extraction with different models.csv',delimiter=';', decimal=',')

plt.style.use('lamalab.mplstyle')


filtered_csv_data = df.dropna(subset=['model', 'number of calls'])

average_calls_csv = filtered_csv_data.groupby('model')['number of calls'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(average_calls_csv['model'], average_calls_csv['number of calls'])
plt.xlabel('Model')
plt.ylabel('Average Number of Calls')
plt.title('Average Number of Calls by Model from CSV Data')

plt.show()
```
